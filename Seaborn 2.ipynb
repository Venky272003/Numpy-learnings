# Categorical Plots

### Categorical Scatter Plots
- StripPlot
- SwarmPlot
### Categorical Distribution Plots
- BoxPlot
- ViolinPlot
### Categorical Estimate Plots -> for central tendency
- BarPlot
- PointPlot
### Figure Level Function -> catplot

```python
import seaborn as sns

# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
# strip plot  (here one column should be categorical and other should be numerical)
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
# jitter
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=False)
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.2)
```

```python
# swarmPlot
sns.swarmplot(data=tips,x='day',y='total_bill',hue="sex")
```

```python
# using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue="sex")
```

```python
# boxplot
sns.boxplot(data=tips,x='sex',y='total_bill',hue="day")
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# catplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='box')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips,y='total_bill')
```

```python
# violinplot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
# hue
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
# barplot
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker')
# When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars
```

```python
# removing the error bars
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',ci=None)
```

```python
# estimator(by default it's mean) --> usimg this we can provide on what basis the bar should be plotted like min,max,values
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=min)
```

```python
# using catplot
sns.catplot(data=tips,x='sex',y='total_bill',hue='smoker',kind='bar')
```

```python
# point plot
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker')
```

```python
#countplot-->A special case for the bar plot is when you want to show the number of observations in each category rather than
#computing a statistic for a second variable.This is similar to a histogram over a categorical,rather than quantitative variable

sns.countplot(data=tips,x='sex',hue='day')
```

```python
# faceting using catplot
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',row='time',kind='box')
```

```python
# Regression plots
---> regplot
---> lmplot
In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x
 and plot the resulting regression line and a 95% confidence interval for that regression.
```

```python
#regplot
# hue parameter not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
#lmplot
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot
sns.residplot(data=tips,x='total_bill',y='tip')
# this plot shows the error in the each point which linear regression made
```

## A second way to plot Facet plots --> FacetGrid

```python
g = sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill')
```

## Plotting Pairwise Relationship (PairGrid vs Pairplot)

```python
sns.pairplot(iris,hue='species')
```

```python
# pairGrid
g = sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
g.add_legend()
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
g.add_legend()
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.add_legend()
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.add_legend()
```
