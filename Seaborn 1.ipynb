# Relational Plots
- to see the statistical relation btw 2 or more variables
- Bivariate Analysis
##### Plots under this section:
- scatter plot
- line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
# Scatter Plot  --> axis level function
sns.scatterplot(data = tips , x = 'total_bill' , y = 'tip',hue='sex',style='time',size='size')
```

```python
# scatter using relplot  ---> figure level function
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
# line plot ---> axis level function
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
sns.lineplot(data = temp_df,x='year',y='lifeExp')
```

```python
# line plot using relplot  ---> figure level fucntion
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# now we plot life expectency of different countries comparing each other
temp_df2 = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df2
```

```python
sns.relplot(kind='line',data=temp_df2,x='year',y='lifeExp',hue='country',style='continent')
```

```python
sns.lineplot(data=temp_df2,x='year',y='lifeExp',hue='country',style='continent')
```

```python
# facet plot  ----> works only with figure level plots ie. relplot
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='smoker',row='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

```python
# col wrap
# first without using col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year')
```

```python
# using col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```

# Distribution Plot
- used for univariate analysis
- used to find out the distribution
- range of the observations
- central tendency
- is the data bimodal
- are there outliers?

#### Plots under distribution plot
- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot,kdeplot,rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# bins parameter
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

```python
# It's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables
sns.displot(data=tips,x='day',kind='hist')

# basically it is countplot but yeah we can say it is histplot
```

```python
# hue parameter 
sns.displot(data=tips,x='tip',kind='hist',hue='sex')
```

```python
# element -> step
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# count of male and female passenger on titanic
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex')
```

```python
# faceting using the col and row   ----> these function will not work for histplot
sns.displot(data=tips,x='tip',kind='hist',element='step',col='sex')
```

```python
# kdeplot
# rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(kind='kde',data=tips,x='total_bill',hue='sex',fill=True)
```

```python
# rugplot
# plot marginal distribution by drawing ticks along the x and y axis
# this function is intended to compliment other plots by showing the location of individual observations in an unobtrusive way

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate Histogram
# A bivariate histogram bins the data within rectangle that title the plot
# and then shows the count of the observations within each rectangle with the fill color
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
# Bivariate kdeplot
# a bivariate kde plot smoothes the (x,y) observations with a 2D Gaussian

sns.kdeplot(data=tips,x='total_bill',y='tip')
```

```python
# Heatmap
# plots rectangular data as a color-encoded matrix
temp_df3 = gap.pivot(index='country',columns='year',values='lifeExp')
# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df3)
```

```python
# annot
temp_df4 = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df4,annot=True,linewidth=0.5,cmap='viridis')
```

```python
# Clustermap
# Plot a matrix dataset as a hierarchically-clustered heatmap
# This function requires scipy to be available
iris= px.data.iris()
iris
```
