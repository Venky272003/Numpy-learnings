```python
batsman = pd.read_csv('sharma-kohli.csv')
print(batsman)
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# Adding title and axis labels
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Career Comparison of Rohit Sharma vs Virat Kohli")
plt.xlabel("Ipl Season")
plt.ylabel("Runs")
```

```python
# changing the color(hex) and the line (width and style) and marker style
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dotted',linewidth=3,marker='D',markersize=15)
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',linewidth=4,marker='o',markersize=8)
plt.title("Career Comparison of Rohit Sharma vs Virat Kohli")
plt.xlabel("Ipl Season")
plt.ylabel("Runs")
```

```python
# adding legend
plt.plot(batsman['index'],batsman['V Kohli'],label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit Sharma')
plt.title("Career Comparison of Rohit Sharma vs Virat Kohli")
plt.xlabel("Ipl Season")
plt.ylabel("Runs")
plt.legend()
```

```python
# we can change the location of the legend

# best
# upper right
# upper left
# lower right
# lower left
# right
# center left
# center rigth
# lower center
# upper center
# center

plt.plot(batsman['index'],batsman['V Kohli'],label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit Sharma')
plt.title("Career Comparison of Rohit Sharma vs Virat Kohli")
plt.xlabel("Ipl Season")
plt.ylabel("Runs")
plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,450000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,100000)  # this shows the graph only for values between 0 to 100000 in the price column
# plt.xlim(2017,2019)
```

```python
# showing gridlines

plt.plot(batsman['index'],batsman['V Kohli'],label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit Sharma')
plt.title("Career Comparison of Rohit Sharma vs Virat Kohli")
plt.xlabel("Ipl Season")
plt.ylabel("Runs")
plt.legend()
plt.grid()
```

```python
# correct way of displaying a graph is using the show() function

plt.plot(batsman['index'],batsman['V Kohli'],label='Virat Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit Sharma')
plt.title("Career Comparison of Rohit Sharma vs Virat Kohli")
plt.xlabel("Ipl Season")
plt.ylabel("Runs")
plt.legend()
plt.grid()
plt.show()
```

### Scatter Plots
- Bivariate Analysis
- Numerical-Numerical data
- Use Case --> Finding Correlation

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('batter.csv')
df = df.head(50)
print(df)
```

```python
plt.scatter(df['avg'],df['strike_rate'])
plt.title('Average and Strike Rate Analysis of top 50 Batsman')
plt.xlabel('average')
plt.ylabel('strike rate')
```

```python
# size  (controlling the size of the marker)
tips = sns.load_dataset('tips')
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
plt.xlabel('total bill')
plt.ylabel('tip')

# here the size of the dot depends on the no of people came along with the customer 
```

```python
# scatterplot using plt.plot
plt.plot(tips['total_bill'],tips['tip'],'o')
# this technique is faster
```

### Bar Chart
- Bivariate Analysis
- Numerical-Categorical data
- Use case --> Aggregate analysis of Groups

```python
# simple bar chart
children = [23,56,43,12,76]
color = ['blue','green','yellow','red','pink']
plt.bar(color,children,color='black')
```

```python
# horizontal bar chart
plt.barh(color,children,color='black')
```

```python
# multiple bar plots
df = pd.read_csv('batsman_season_record.csv')
print(df)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)
```

```python
# xticks
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2,label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2,label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2,label='2017')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.legend()
```

```python
# A problem
children = [23,56,43,12,76]
color = ['blue blueblueblueblue','green green green green green','yellow yellow yellow yellow','red red red red','pink pink pink']
plt.bar(color,children,color='black')
```

```python
# solution
children = [23,56,43,12,76]
color = ['blue blueblueblueblue','green green green green green','yellow yellow yellow yellow','red red red red','pink pink pink']
plt.bar(color,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# stacked bar chart 
df = pd.read_csv('batsman_season_record.csv')

plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']),label='2015')
plt.legend()
```

### Pie Chart
- Univariate/Bivariate Analysis
- Categorical-Numerical
- Use Case --> To find the contribution on a standard scale

```python
# simple data
data = [23,45,67,89,100]
subjects = ['English','Maths','Hindi','SST','Science']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# on dataset
df = pd.read_csv('gayle-175.csv')
print(df)
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# changing colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','yellow','green','red','pink','purple'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
plt.savefig('batsman.png')
```
