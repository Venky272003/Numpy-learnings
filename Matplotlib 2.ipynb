```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal length in cm')
plt.ylabel('Petal length in cm')
```

```python
# now we changeing the values in the species field like this
# setosa --> 0
# versicolor --> 1
# virginica --> 2
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('Sepal length in cm')
plt.ylabel('Petal length in cm')
```

```python
# we can change the color also using the cmap
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('Sepal length in cm')
plt.ylabel('Petal length in cm')
```

```python
# adding colorbar
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('Sepal length in cm')
plt.ylabel('Petal length in cm')
plt.colorbar()
```

```python
# changing the opacity of the markers using alpha attribute
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.2)
plt.xlabel('Sepal length in cm')
plt.ylabel('Petal length in cm')
plt.colorbar()
```

## Changing the plot size

```python
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('Sepal length in cm')
plt.ylabel('Petal length in cm')
plt.colorbar()
```

## Annotations


```python
batters = pd.read_csv('batter.csv')
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state = 5)
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i],fontdict={'size':12,'color':'red'})
    
# annotation means we are giving a name to each point
# the size of each point is based on the runs scored by the batsman
```

### Horizontal and Vertical Lines

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

#**************************************************************************************
plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')
#**************************************************************************************
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## Subplots
- we can plot multiple graphs side by side


```python
batters.head()
```

```python
fig,ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'],batters['strike_rate'],color='yellow',marker='D')
ax.set_title('Avg vs Strike Rate Comparison')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
# plotting multiple graphs

fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='orange')
ax[1].scatter(batters['avg'],batters['runs'],color='green')

ax[0].set_title('Average vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Average vs Total Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
```

```python
# plotting 4 graphs

fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='orange')
ax[0,1].scatter(batters['avg'],batters['runs'],color='green')
ax[1,0].hist(batters['avg'],color='red')
ax[1,1].hist(batters['runs'],color='yellow')
```

```python
# Another way of plotting subplots
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)  # 2 rows 1 column 1st graph
ax1.scatter(batters['avg'],batters['strike_rate'],color='orange')

ax2 = fig.add_subplot(2,1,2)  # 2 rows 1 column 2nd graph
ax2.hist(batters['runs'],color='yellow')
```

## 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
# we directly cannot add colorbar here instead we do this by storing the plot in a variable and then passing the variable in the colobar()
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# second graph
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Contour Plots

```python
fig = plt.figure()
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# countour filled
fig = plt.figure()
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Heatmap

```python
delivery = pd.read_csv('IPL_EveryBall.csv')
delivery.head()
```

```python
only_six = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
only_six
```

```python
grid = only_six.pivot_table(index = 'overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
```

```python
plt.figure(figsize=(20,8))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))

plt.colorbar()
```

```python
# pie chart
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)
df.head()
```
